## Create groups of 65 slices

```python
in_path = r'C:\Users\Admin\Documents\Uni\DS200\liver_segmentation\datasets\dicom_files\images'
out_path = r'C:\Users\Admin\Documents\Uni\DS200\liver_segmentation\datasets\dicom_groups\images'
```

```python
from glob import glob
import shutil
import os
import dicom2nifti
import nibabel as nib
import numpy as np
```

```python
patient_list = glob(in_path + '\*')
print(patient_list[0])
```

```python
for patient in glob(in_path + '/*'):
    
    patient_name = os.path.basename(os.path.normpath(patient))
    
    files = sorted(glob(patient + '/*'))
    
    number_folders = int(len(files) / 65)

    for folder_idx in range(number_folders):

        output_path_name = os.path.join(out_path, patient_name + '_' + str(folder_idx))
        os.mkdir(output_path_name)

        start = folder_idx * 65
        end = start + 65

        for file in files[start:end]:
            shutil.move(file, output_path_name)
```

```python
in_path_images = r'C:\Users\Admin\Documents\Uni\DS200\liver_segmentation\datasets\dicom_groups\images\*'
in_path_labels = r'C:\Users\Admin\Documents\Uni\DS200\liver_segmentation\datasets\dicom_groups\labels\*'
out_path_images = r'C:\Users\Admin\Documents\Uni\DS200\liver_segmentation\datasets\nifti_files\images'
out_path_labels = r'C:\Users\Admin\Documents\Uni\DS200\liver_segmentation\datasets\nifti_files\labels'

list_images = glob(in_path_images)
list_labels = glob(in_path_labels)
```

```python
list_images
```

```python
for patient in list_images:
    patient_name = os.path.basename(os.path.normpath(patient))
    dicom2nifti.dicom_series_to_nifti(patient, os.path.join(out_path_images, patient_name + '.nii.gz'))
```

```python
for patient in list_labels:
    patient_name = os.path.basename(os.path.normpath(patient))
    dicom2nifti.dicom_series_to_nifti(patient, os.path.join(out_path_labels, patient_name + '.nii.gz'))
```
